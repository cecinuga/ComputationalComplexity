```python
import matplotlib.pyplot as plt
from algorithms.insertion_sort import insertion_sort
from algorithms.merge_sort import merge_sort
from algorithms.bubble_sort import bubble_sort
from algorithms.heap_sort import heap_sort
from algorithms.quick_sort import quick_sort, quick_sort_randomized, quick_sort_median
from analyzer import generate__list_of_naturals__test_cases
```

#### Perform Sorting

```python
n_test_case = 30

algorithms_test_cases = {
    'bubble_sort': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': bubble_sort, 'complexity': 'n^2' },
    'quick_sort': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': quick_sort, 'complexity': 'n^2' },
    'insertion_sort':{ 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': insertion_sort, 'complexity':'n^2' }, 
    'heap_sort': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': heap_sort, 'complexity': 'n logn' },
    'quick_sort_median': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': quick_sort_median, 'complexity': 'n logn' },
    'quick_sort_randomized': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': quick_sort_randomized, 'complexity': 'n logn' },
    'merge_sort': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': merge_sort, 'complexity': 'n logn' },
}

for algo in algorithms_test_cases:
    for test_case in algorithms_test_cases[algo]['test']['test_case']:
        test_case.read_counter = 0
        algorithms_test_cases[algo]['algo'](test_case)
```

```python
plt.figure(figsize=(10, 6))
plt.ylabel('Operations')
plt.xlabel('Input Length: n')

for algo in algorithms_test_cases:
    plt.plot(
        [ i for i in algorithms_test_cases[algo]['test']['length'] ], 
        [ test_case.read_counter for test_case in algorithms_test_cases[algo]['test']['test_case'] ],
        label=r'{0} -{1}'.format(algo, r'$\ O({0})$'.format(algorithms_test_cases[algo]['complexity']))
    ) 
      
plt.legend() 
plt.title("Comparison method")
plt.savefig('source/sorting_complexity.png')
plt.show()
```
