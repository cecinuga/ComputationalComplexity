```python
import matplotlib.pyplot as plt
from algorithms.cut_rod_topdown import cut_rod_topdown
from algorithms.cut_rod_mem_topdown import cut_rod_mem_topdown
from algorithms.cut_rod_bottomup import cut_rod_bottomup
from analyzer import generate__list_of_random_naturals__test_cases
```

#### Perform Dynamic Stuffsss

```python
n_test_case = 8

algorithms_test_cases = {
    'cut_rod_topdown': { 'test': generate__list_of_random_naturals__test_cases(n_test_case), 'algo': cut_rod_topdown, 'complexity': '2^n' },
    'cut_rod_mem_topdown': { 'test': generate__list_of_random_naturals__test_cases(n_test_case), 'algo': cut_rod_mem_topdown, 'complexity': 'n^2' },
    'cut_rod_mem_topdown': { 'test': generate__list_of_random_naturals__test_cases(n_test_case), 'algo': cut_rod_mem_topdown, 'complexity': 'n^2' },
    'cut_rod_bottomup': { 'test': generate__list_of_random_naturals__test_cases(n_test_case), 'algo': cut_rod_bottomup, 'complexity': 'n^2' },
}

for algo in algorithms_test_cases:
    for test_case in algorithms_test_cases[algo]['test']['test_case']:
        test_case.read_counter = 0
        algorithms_test_cases[algo]['algo'](test_case)
```

```python
plt.figure(figsize=(10, 6))
plt.ylabel('Operations')
plt.xlabel('Input Length: n')
for algo in algorithms_test_cases:
    plt.plot(
        [ i for i in algorithms_test_cases[algo]['test']['length'] ], 
        [ test_case.read_counter for test_case in algorithms_test_cases[algo]['test']['test_case'] ],
        label=r'{0} -{1}'.format(algo, r'$\ O({0})$'.format(algorithms_test_cases[algo]['complexity']))
    ) 
      
plt.legend()  
plt.title("Dynamic Shit")
plt.savefig('source/dynamic_programming.png')
plt.show()
```
