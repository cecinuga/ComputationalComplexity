```python
import matplotlib.pyplot as plt
from analyzer import generate__list_of_naturals__test_cases, generate__list_of_non_sorted_integers__test_case
from algorithms.linear_search import linear_search
from algorithms.binary_search import binary_search
from algorithms.max_sub_array import max_sub_array
```

```python
n_test_case = 100

algorithms_test_cases = {
    'linear_search': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': linear_search, 'complexity': 'n' }, 
    'binary_search': { 'test': generate__list_of_naturals__test_cases(n_test_case), 'algo': binary_search, 'complexity': 'logn' },
    'max_sub_array': { 'test': generate__list_of_non_sorted_integers__test_case(n_test_case), 'algo': max_sub_array, 'complexity': 'n^2' },
}

for algo in algorithms_test_cases:
    for test_case in algorithms_test_cases[algo]['test']['test_case']:
        test_case.read_counter = 0
        if algo == 'max_sub_array':
            algorithms_test_cases[algo]['algo'](test_case)
            continue

        algorithms_test_cases[algo]['algo'](test_case, float('inf'))
```

```python
plt.figure(figsize=(10, 6))
plt.ylabel('Operations')
plt.xlabel('Input Length: n')

for algo in algorithms_test_cases:
    plt.plot(
        [ i for i in algorithms_test_cases[algo]['test']['length'] ], 
        [ test_case.read_counter for test_case in algorithms_test_cases[algo]['test']['test_case'] ],
        label=r'{0} -{1}'.format(algo, r'$\ {0}$'.format(algorithms_test_cases[algo]['complexity']))
    )

plt.legend()
plt.savefig('source/searching_complexity.png')
plt.show()
```
